```python
import math
import csv
def load_csv(filename):
        lines = csv.reader(open(filename, "r"));
        dataset = list(lines)
        headers = dataset.pop(0)
        return dataset, headers
```

```python
class Node:
    def __init__(self, attribute):
        self.attribute = attribute
        print("self.attribute")
        print(self.attribute)
        self.children = []
        print(self.children)
        self.answer = ""   # NULL indicates children exists. # Not Null indicates this is a Leaf Node
     # print(self.answer)
```

```python
def subtables(data, col, delete):
    dic = {}
    coldata = [ row[col] for row in data]
    attr = list(set(coldata)) # All unique values of attribute retrived
    print("attr")
    print(attr)
    for k in attr:
        dic[k] = []
        print(dic[k])    
    for y in range(len(data)):#len=14 ,0 to 13
        key = data[y][col]
        print(key)
        if delete:
            del data[y][col]
        dic[key].append(data[y])
        #print("dic[key]")
        print(dic) 

    return attr, dic
```

```python
def entropy(S):
    print("entropy")  
    attr = list(set(S))
    print(attr)
    if len(attr) == 1: #if all are +ve/-ve then entropy = 0 
        return 0
    counts = [0,0] # Only two values possible 'yes' or 'no' 
    for i in range(2):
        counts[i] = sum( [1 for x in S if attr[i] == x] ) / (len(S) * 1.0)
        print(counts[i])
    sums = 0
    for cnt in counts:
        sums += -1 * cnt * math.log(cnt, 2)
    return sums
```

```python
def compute_gain(data, col):

    attValues, dic = subtables(data, col, delete=False)
    print(attValues)
    print(dic)
    print([row[-1] for row in data])# last col value
    total_entropy = entropy([row[-1] for row in data])
    print("out total_entropy")
    print(total_entropy)
    for x in range(len(attValues)):

        ratio = len(dic[attValues[x]]) / ( len(data) * 1.0) 
        print("ratio")# rainy=5/14 =0.35
        print(ratio)
        entro = entropy([row[-1] for row in dic[attValues[x]]]) 
        print(entro) #0.97
        total_entropy -= ratio*entro
        print("in total_entropy")
        print(total_entropy)
    return total_entropy
```

```python
def build_tree(data, features):
    
    lastcol = [row[-1] for row in data]
    
    print(lastcol)
    
    print(len(set(lastcol)))

    print("hi")
    
    if (len(set(lastcol))) == 1: # If all samples have same labels return that label 
        node=Node("")
        print(node.attribute)
        print(node.children)
        node.answer = lastcol[0]
        print("hello")
        return node

    print(data[0])#it gives all the values in a row from which col count can be obtained
    n = len(data[0])-1
    print(n)
    
    gains = [compute_gain(data, col) for col in range(n) ]
    print("GAINS")
    print(gains)
    #print("end")

    split = gains.index(max(gains)) # Find max gains and returns index 
    #print("split")
    print(split)   # index  0 ,..
    node = Node(features[split]) # 'node' stores attribute selected 
#del (features[split])
    print("node")
    print(node)
    fea = features[:split]+features[split+1:]
    print("fea")
    print(fea)
    attr, dic = subtables(data, split, delete=True) # Data will be spilt in subtables 
    
    print(attr) 
    print(dic)
    for x in range(len(attr)):
        child = build_tree(dic[attr[x]], fea)

        node.children.append((attr[x], child))

    return node
```

```python
def print_tree(node, level):

    if node.answer != "":#if child does not exist then print yes/no

        print(" "*level, node.answer) # Displays leaf node yes/no
        return
    print(" "*level, node.attribute) # Displays attribute Name 
    for value, n in node.children:
        print("	"*(level+1), value)
        print_tree(n, level + 2)
```

```python
def classify(node,x_test,features):

    if node.answer != "":  #child does not exist. It is the leaf node

        print(node.answer)

        return

    pos = features.index(node.attribute)
    print(pos)

    for value, n in node.children:

        if x_test[pos]==value:

            classify(n,x_test,features)

''' Main program '''

dataset, features = load_csv("tennis12.csv") # Read Tennis data 
node = build_tree(dataset, features) # Build decision tree

print("The decision tree for the dataset using ID3 algorithm is ")

print_tree(node, 0)

testdata, features = load_csv("tennis12.csv")

for xtest in testdata:

    print("The test instance : ",xtest)
    print("The predicted label : ",end="")
    classify(node,xtest,features)
```
